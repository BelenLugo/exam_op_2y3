```python
#Vemos que los datos duplicados pueden contener datos utilizables en un futuro, para ellos usamos df2 antes de eliminarlos
df2.drop_duplicates(inplace=True)
df2.duplicated().sum()
```

```python
# Definir el diccionario de tipos de datos
diccionario = {
    'hotel': 'string',
    'is_canceled': 'int64',
    'lead_time': 'int64',
    'arrival_date_year': 'int64',
    'arrival_date_month': 'string',
    'arrival_date_week_number': 'int64',
    'arrival_date_day_of_month': 'int64',
    'stays_in_weekend_nights': 'int64',
    'stays_in_week_nights': 'int64',
    'adults': 'int64',
    'children': 'float64',  # La columna contiene NA
    'babies': 'int64',
    'meal': 'string',
    'country': 'string',
    'market_segment': 'string',
    'distribution_channel': 'string',
    'is_repeated_guest': 'int64',
    'previous_cancellations': 'int64',
    'previous_bookings_not_canceled': 'int64',
    'reserved_room_type': 'string',
    'assigned_room_type': 'string',
    'booking_changes': 'int64',
    'deposit_type': 'string',
    'agent': 'float64',  # La columna contiene NA
    'company': 'float64',  # La columna contiene NA
    'days_in_waiting_list': 'int64',
    'customer_type': 'string',
    'adr': 'float64',
    'required_car_parking_spaces': 'int64',
    'total_of_special_requests': 'int64',
    'reservation_status': 'string',
    'reservation_status_date': 'string',
}

# Verificar y ajustar tipos de datos
def verificar_y_ajustar_tipos(df2, diccionario):
    for columna, tipo_esperado in diccionario.items():
        if columna in df2.columns:
            try:
                # Ajustar el tipo de datos
                df2[columna] = df2[columna].astype(tipo_esperado)
            except Exception as e:
                print(f"Error al convertir la columna '{columna}' a '{tipo_esperado}': {e}")

verificar_y_ajustar_tipos(df2, diccionario)
print("Tipos de datos en el DataFrame:")
print(df2.dtypes)
```

```python
columnas_categoricas = ['meal', 'country', 'market_segment', 'customer_type', 'reservation_status']
for columna in columnas_categoricas:
    print(f"Valores únicos en '{columna}':")
    print(df2[columna].unique())
    print("\n")
```

```python
# Normalizar valores a minúsculas
df2['meal'] = df2['meal'].str.lower()
df2['country'] = df2['country'].str.lower()
df2['market_segment'] = df2['market_segment'].str.lower()
df2['customer_type'] = df2['customer_type'].str.lower()
df2['reservation_status'] = df2['reservation_status'].str.lower()

# Reemplazos para la columna 'meal'
meal_replacements = {
    'bb': 'bed & breakfast',
    'fb': 'full board',
    'hb': 'half board',
    'sc': 'self catering',
    'undefined': 'no meal'
}

df2['meal'] = df2['meal'].replace(meal_replacements)

# Reemplazos para la columna 'market_segment'
market_segment_replacements = {
    'undefined': 'unknown'
}

df2['market_segment'] = df2['market_segment'].replace(market_segment_replacements)

# Para 'country': podrías usar nombres completos si es necesario, pero aquí mantenemos las abreviaturas.
# Para esto, no haremos cambios, a menos que se requiera un reemplazo específico.

# Comprobamos los resultados
print("Valores únicos en 'meal' después de la normalización:")
print(df2['meal'].unique())
print("\nValores únicos en 'country' después de la normalización:")
print(df2['country'].unique())
print("\nValores únicos en 'market_segment' después de la normalización:")
print(df2['market_segment'].unique())
print("\nValores únicos en 'customer_type' después de la normalización:")
print(df2['customer_type'].unique())
print("\nValores únicos en 'reservation_status' después de la normalización:")
print(df2['reservation_status'].unique())
```

```python
# Identificamos los valores faltantes
valores_faltantes = df2.isnull().sum()
print("Valores faltantes por columna:")
print(valores_faltantes[valores_faltantes > 0])
```

```python
# 1. Convertir 'agent' y 'company' a tipo string
df['agent'] = df['agent'].astype(str)
df['company'] = df['company'].astype(str)

# 3. Rellenar 'country', 'agent' y 'company' con 'Desconocido'
df['country'] = df['country'].fillna('Desconocido')
df['agent'] = df['agent'].fillna('Desconocido')
df['company'] = df['company'].fillna('Desconocido')
```

```python
valores_faltantes = df2.isnull().sum()
print("Valores faltantes por columna:")
print(valores_faltantes[valores_faltantes > 0])
```

```python
# Rellenar valores faltantes con 0 para children y desconocido para country
df2.loc[df2['children'].isnull(), 'children'] = 0
df2.loc[df2['country'].isnull(), 'country'] = 'Desconocido'

# Verificar que no haya más valores nulos
print("Valores faltantes después de imputar:")
print(df2.isnull().sum())
```

```python
# Calcular IQR para identificar valores atípicos
Q1 = df2['stays_in_week_nights'].quantile(0.25)
Q3 = df2['stays_in_week_nights'].quantile(0.75)
IQR = Q3 - Q1

# Definir límites superior e inferior
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

# Identificar los outliers
outliers = df2[(df2['stays_in_week_nights'] < lower_bound) | (df2['stays_in_week_nights'] > upper_bound)]
print("Datos anómalos en 'stays_in_week_nights':")
print(outliers)
```

```python
# Visualización univariada: Histograma de 'stays_in_week_nights'
plt.figure(figsize=(10, 6))
sns.histplot(df2['stays_in_week_nights'], bins=30, kde=True, color='skyblue')
plt.title('Distribución de Estancias en Noches')
plt.xlabel('Noches de Estancia')
plt.ylabel('Frecuencia')
plt.axvline(df2['stays_in_week_nights'].mean(), color='red', linestyle='dashed', linewidth=1, label='Media')
plt.axvline(df2['stays_in_week_nights'].median(), color='purple', linestyle='dashed', linewidth=1, label='Mediana')
plt.legend()
plt.show()
```

##### La distribución de estancias en noches muestra que la mayoría de las estancias son de corta duración, con una concentración en las estancias de 1 a 7 noches. Sin embargo, hay algunas estancias más largas que son menos frecuentes.

```python
# Visualización univariada: Boxplot de 'adr' (precio diario medio) por tipo de hotel
plt.figure(figsize=(10, 6))
sns.boxplot(x='hotel', y='adr', data=df, palette='pastel')
plt.title('Boxplot del Precio Diario Medio (ADR) por Tipo de Hotel')
plt.xlabel('Tipo de Hotel')
plt.ylabel('Precio Diario Medio (ADR)')
plt.xticks(rotation=15) 
plt.show()
```

#### Este boxplot muestra la distribución del precio diario medio (ADR) para cada tipo de hotel. Puedes observar la mediana, los cuartiles y cualquier valor atípico. Esto te permitirá entender cómo varían los precios según el tipo de hotel.

```python
# Visualización multivariada: Gráfico de dispersión entre 'lead_time' y 'adr'
plt.figure(figsize=(10, 6))
sns.scatterplot(x='lead_time', y='adr', hue='hotel', data=df2, alpha=0.6)
plt.title('Relación entre el Tiempo de Anticipación y el Precio Diario Medio')
plt.xlabel('Tiempo de Anticipación (Lead Time)')
plt.ylabel('Precio Diario Medio (ADR)')
plt.legend(title='Tipo de Hotel')
plt.show()
```

#### Este gráfico de dispersión sugiere que hay una relación positiva entre el tiempo de anticipación y el precio diario medio, con los 'Resort Hotels' mostrando precios más altos en comparación con los 'City Hotels', especialmente en tiempos de anticipación más largos.

```python
# Visualización multivariada: distribución del lead_time según el tipo de hotel y el estado de cancelación.

plt.figure(figsize=(12, 8))
sns.boxplot(x='hotel', y='lead_time', hue='is_canceled', data=df2, palette='pastel')
plt.title('Distribución del Lead Time por Tipo de Hotel y Estado de Cancelación')
plt.xlabel('Tipo de Hotel')
plt.ylabel('Lead Time (Días)')
plt.xticks(rotation=15)
plt.legend(title='¿Reserva Cancelada? (1 = Sí, 0 = No)')
plt.show()
```

2. Identificación de Tendencias
A. Tasa de Cancelaciones
La tasa de cancelación del 27.5% es notable y podría indicar factores que afectan la decisión de los huéspedes. Se puede investigar más a fondo los motivos de cancelación, como cambios en políticas, eventos inesperados, o fluctuaciones en la demanda.

B. Comportamiento de Reservas
Anticipación de Reservas: La amplia variabilidad en el lead_time sugiere que algunos huéspedes planean con mucha anticipación, mientras que otros hacen reservas de última hora. Este comportamiento puede estar influenciado por la época del año, promociones o la naturaleza del viaje.

Estancias Cortas: El promedio de estancias (1 fin de semana y 2 noches en total) podría ser un indicativo de que los huéspedes tienden a realizar viajes cortos, lo cual puede ser relevante para el marketing y las estrategias de precios.

C. Comportamiento de los Huéspedes
La baja tasa de huéspedes recurrentes sugiere que el hotel podría beneficiarse de programas de fidelización o promociones dirigidas a clientes que ya han reservado anteriormente.
Con solo un promedio de 1.88 adultos por reserva, se puede deducir que muchos huéspedes están viajando solos o en parejas, lo que podría impactar en las decisiones sobre el tipo de habitación o las ofertas de servicios.


Este análisis ofrece una perspectiva sobre cómo se comportan las reservas y los huéspedes. Las tendencias identificadas pueden ser aprovechadas para afinar las estrategias de marketing, ajustar precios y enriquecer la experiencia del cliente. Comprender estos patrones es fundamental para aumentar la tasa de ocupación y reducir las cancelaciones.
